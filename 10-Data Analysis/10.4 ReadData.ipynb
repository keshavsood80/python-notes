Read Data From Different Sources

```python
# by using some Inbuilt functions that are available in pandas
import pandas as pd
from io import StringIO
Data = '{"employee_name": "James", "email": "james@gmail.com", "job_profile": [{"Title1": "Team Lead","Title2": "Sr. Developer"}]}'
# Data is in the form of Json
# I will convert it into String by using StringIO
df = pd.read_json(StringIO(Data))
```

```python
# json Data can convert into DataFrame
df
```

```python
# DataFrame can convert back into json
df.to_json()
```

```python
# DataFrame can convert back into json based on records
df.to_json(orient='index')
```

```python
df.to_json(orient='records')
```
