```python
import seaborn as sns
```

```python
# Basic Plotting with Seaborn
tips = sns.load_dataset('tips')
tips # inbuilt datasets are available
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(data=tips,x='total_bill',y='tip')
plt.title('Scatter Plot of total bill vs tip')
plt.show()
```

```python
# Line Plot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title('Line Plot of total bill by size')
plt.show()
```

```python
# Categorical Plots
# bar Plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar Plot of Total bill by day')
plt.show()
```

```python
# Box Plot 
# Box Plot is also an amazing way to seeing outliers.
sns.boxplot(data=tips,x='total_bill',y='day')
```

```python
# Violin Plot
# to see the distribution of the data
sns.violinplot(data=tips,x='day',y='total_bill')
```

```python
# Histogram

sns.histplot(tips['total_bill'],bins=10,kde=True)# kde means kernel density estimator -True means that I get the curve.
```

```python
# KDE Plot
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
# Pair plot
# to see relationship b/w all the variable to the different -2 variables.
# Numerical variable
sns.pairplot(tips)
```

```python
# Heatmap
# to see the correlation b/w all the numerical variables

corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
