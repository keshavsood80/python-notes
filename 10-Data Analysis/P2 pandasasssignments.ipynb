```python
### Assignment 3: Data Cleaning

#1. Create a Pandas DataFrame with 3 columns and 5 rows filled with random integers.
# Introduce some NaN values. Fill the NaN values with the mean of the respective columns.


df= pd.DataFrame(np.random.randint(1,21,size =(5,3)),columns=['A','B','C'])
print('Original Dataframe')
print(df)

# Introduce some NaN values
df.iloc[0,1]= np.nan
df.at[2,'A']= np.nan
df.at[4,'C']= np.nan
print('Dataframe with Nan values')
print(df)

# Fill the NaN values with the mean of the respective columns
df.fillna(df.mean(),inplace=True)
print('Dataframe with nan values filled ')
df
```

```python
### Assignment 4: Data Aggregation

# 1. Create a Pandas DataFrame with 2 columns: 'Category' and 'Value'.
# Fill the 'Category' column with random categories ('A', 'B', 'C') and
# the 'Value' column with random integers. Group the DataFrame by 'Category' and
#  compute the sum and mean of 'Value' for each category.
df = pd.DataFrame({'Category': np.random.choice(['A','B','C'],size=10),'Value':np.random.randint(1,101,size=10)})
print('Original Dataframe:')
print(df)
print('Grouped Dataframe:')
df.groupby('Category')['Value'].agg(['sum','mean'])
```

```python
### Assignment 5: Merging DataFrames

# 1. Create two Pandas DataFrames with a common column.
#  Merge the DataFrames using the common column.
df1 = pd.DataFrame({'key':['A','B','C','D'],'Value':[1,2,3]})
df2 = pd.DataFrame({'key':['A','B','E','D'],'Value':[4,5,6]})
print('Dataframe 1:')
print(df1)
print('Dataframe 2:')
print(df2)
print('Merged Dataframe:')
merged = pd.merge(df1,df2,on='key',how='inner')
print(merged)
```

```python
# 2. Create two Pandas DataFrames with different columns.
#  Concatenate the DataFrames along the rows and along the columns.
df1 = pd.DataFrame({'A':np.random.randint(1,111,size=4),'B':np.random.randint(1,111,size=4)})
df2 = pd.DataFrame({'C':np.random.randint(1,111,size=4),'D':np.random.randint(1,111,size=4)})
print('Dataframe 1:')

print(df1)
print('Dataframe 2:')
print(df2)

rows = pd.concat([df1,df2],axis=0)
cols = pd.concat([df1,df2],axis=1)
print('Concatenated Dataframe (rows):')
print(rows)
print('Concatenated Dataframe (cols):')
print(cols)
```

```python
### Assignment 6: Time Series Analysis

# 1. Create a Pandas DataFrame with a datetime index and one column filled with random integers.
#  Resample the DataFrame to compute the monthly mean of the values.


date_rng= pd.date_range(start='1-1-2024',end='12-12-2024',freq='D')
df = pd.DataFrame(date_rng,columns=['date'])
df['data']= np.random.randint(111,1111,size=(len(date_rng)))
df.set_index('date',inplace=True)
print('Original Df:')
print(df)

print('mean_dataframe:')
df.resample('ME').mean()
```

```python
# 2. Create a Pandas DataFrame with a datetime index ranging from '2021-01-01' to '2021-12-31' and
#  one column filled with random integers. Compute the rolling mean with a window of 7 days.

date_rng = pd.date_range(start='2021-01-01',end= '2021-12-31',freq='D')
df = pd.DataFrame(date_rng,columns=['date'])
df['data']= np.random.randint(111,1111,size=len(date_rng))
df.set_index('date',inplace=True)
print('df:')
print(df)

rolling_mean = df.rolling(window=7).mean()
print('rolling_mean:')
rolling_mean
```

```python
### Assignment 8: Pivot Tables

# 1. Create a Pandas DataFrame with columns 'Date', 'Category', and 'Value'. 
# Create a pivot table to compute the sum of 'Value' for each 'Category' by 'Date'.
date_rng = pd.date_range(start='2024-12-1',end='2024-12-10',freq='D')
df = pd.DataFrame({'Date':np.random.choice(date_rng,size=20) ,'Category': np.random.choice(['A','B','C'],size=20),'Value': np.random.randint(11,1111,size = 20)})
print('Original Dataframe:')
print(df)

pivot_table = pd.pivot_table(df,values = 'Value',index='Date',columns='Category',aggfunc='sum')
print('Pivot table:')
pivot_table
```

```python
# 2. Create a Pandas DataFrame with columns 'Year', 'Quarter', and 'Revenue'.
#  Create a pivot table to compute the mean 'Revenue' for each 'Quarter' by 'Year'.
df = pd.DataFrame({'Year':np.random.choice(['2021','2022','2023','2024'],size=12),'Quarter':np.random.choice(['Q1','Q2','Q3','Q4'],size=12),'Revenue':np.random.randint(11,111,size=12)})
print('Original DataFrame:')
print(df)
pivot_table= pd.pivot_table(df,values='Revenue',index='Year',columns='Quarter',aggfunc='mean')
pivot_table
```

```python
### Assignment 9: Applying Functions

#1. Create a Pandas DataFrame with 3 columns and 5 rows filled with random integers.
# Apply a function that doubles the values of the DataFrame.
df = pd.DataFrame(np.random.randint(11,111,size=(5,3)))
print('Original df:')
print(df)
double_df = df.apply(lambda x: x*2)
print('double_df:')
double_df
```
